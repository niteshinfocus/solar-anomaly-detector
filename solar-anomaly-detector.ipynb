```python
# Install YOLOv8 (if you haven't already in this session)
!pip install -q ultralytics

import os
from ultralytics import YOLO

# 1. Define the exact path kagglehub gave you
# (Using the variable from your previous cell, or pasting the string directly)
dataset_path = "/root/.cache/kagglehub/datasets/pkdarabi/solarpanel/versions/49"

# 2. Locate the required data.yaml file dynamically
yaml_path = None
for root, dirs, files in os.walk(dataset_path):
    if 'data.yaml' in files:
        yaml_path = os.path.join(root, 'data.yaml')
        break

if yaml_path:
    print(f"Success! Found YAML configuration at: {yaml_path}")

    # 3. Initialize the YOLOv8 'nano' model (fastest for initial testing)
    model = YOLO('yolov8n.pt')

    # 4. Begin training
    print("Starting YOLOv8 training...")
    results = model.train(
        data=yaml_path,
        epochs=30,               # You can increase this later for better accuracy
        imgsz=640,               # Native resolution of the dataset
        batch=16,                # Number of images processed at once
        project='SolarAnomaly',  # Output folder name
        name='yolov8_run'        # Sub-folder for this specific run
    )
else:
    print("Error: data.yaml not found. Let's double-check the path contents.")
```

```python
import os
from ultralytics import YOLO
from google.colab import files

# 1. Load your trained model using the EXACT path from your training log
model_path = '/content/runs/detect/SolarAnomaly/yolov8_run/weights/best.pt'
model = YOLO(model_path)

# 2. Export to standard ONNX format (Full Accuracy, FP32 precision)
model.export(format='onnx')

# 3. The export function creates the .onnx file in the exact same folder
onnx_path = '/content/runs/detect/SolarAnomaly/yolov8_run/weights/best.onnx'

if os.path.exists(onnx_path):
    print("Export successful! Triggering download...")
    files.download(onnx_path)
else:
    print("Error: Could not find the generated ONNX file. Let's check the folder.")
```
