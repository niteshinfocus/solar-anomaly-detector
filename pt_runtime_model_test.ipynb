```python
!pip install ultralytics -q

from ultralytics import YOLO
from google.colab.patches import cv2_imshow # Colab's special display function

# 1. Load model
model = YOLO("best.pt")

# 2. Run inference
# (Replace "my_actual_image.jpg" with your real file name!)
results = model.predict(source="test-img2.jpg", save=True)

# 3. Display the predicted image in the notebook
# YOLOv8 saves outputs by default to 'runs/detect/predict/'

# 4. Display the result directly from memory
# .plot() generates the image with bounding boxes, and cv2_imshow displays it
cv2_imshow(results[0].plot())
```
