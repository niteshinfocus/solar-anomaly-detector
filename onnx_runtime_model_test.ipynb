```python
!pip install ultralytics onnxruntime onnx -q

from ultralytics import YOLO

# 1. Load your model (works for both .pt or .onnx)
model = YOLO("best.onnx", task="detect")

# 2. Run inference on the entire folder and save to a custom location
results = model.predict(
    source="test-images/", # Point this to your folder of input images
    save=True,                    # Tell YOLO to save the annotated images
    project="/content/",          # The base directory (Colab's default file explorer tab)
    name="my_results",            # The name of the folder where results will be saved
    exist_ok=True                 # Prevents YOLO from making 'my_results2', 'my_results3', etc.
)

print("Inference complete! Check the 'my_results' folder on the left.")

#run it seperately as bash file if you have multiple images in result
#!zip -r /content/my_results.zip /content/my_results
```
